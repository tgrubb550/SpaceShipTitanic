```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv(r"../Data/train.csv")
```

```python
data.describe()
```

```python
data.info()
```

```python
data.columns
```

```python
Disc_Data = ['HomePlanet', 'CryoSleep', 'Destination','VIP']
Cont_Data = ['Age','RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
```

```python
for d in Disc_Data:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x=d, hue='Transported')
    plt.xlabel(d)
    plt.ylabel('Frequency')
    plt.title(f'Bar Chart of {d}')
    plt.show()
```

```python
for d in Cont_Data:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x=d, bins=30, hue='Transported', kde=False, alpha=0.6)  
    plt.xlabel(d)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {d} by Transported')
    plt.legend(title='Transported')
    plt.show()
```
